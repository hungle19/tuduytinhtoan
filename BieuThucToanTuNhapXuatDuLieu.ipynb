bai 3: Viet chuong trinh nhan vao hai so nguyen a, btu ban phim. In ra tong, hieu, tich, phan nguyen, phan du va ket qua thuc hien phep chia a cho b (lay den 2 chu so phan thap phan) theo dinh dang dung cua output 

```python
a = int(input("10"))
b = int(input("3"))
tong = a + b
hieu = a - b
tich = a * b
phan_nguyen = a // b
phan_du = a % b
phep_chia = a / b
print (f"tong: {tong}")
print (f"hieu: {hieu}")
print (f"tich: {tich}")
print (f"phan_nguyen: {phan_nguyen}")
print (f"phan_du: {phan_du}")
print (f"phep_chia: {phep_chia:.2f}")
```

bai 4: Viết chương trình python để tính điểm trung bình môn học của một học sinh trong lớp. Có tổng cộng 6 đầu điểm: a1, b1, c1, a2, b2, a3. Trong đó:
• Các điểm a1, b1, c1 được nhân hệ số 1,
• Các điểm a2, b2 được nhân hệ số 2,
• Điểm a3 được nhân hệ số 3.
Công thức tính điểm trung bình là: TB=((a1+b1+c1)+(a2+b2)×2+a3×3)/10
Đầu vào: Trong bài tập này, dữ liệu vào chỉ gồm duy nhất một dòng bao gồm 6 số thể hiện cho 6 đầu điểm a1, b1, c1, a2, b2, a3 theo đúng thứ tự đó. Trong đó, hai đầu điểm liên tiếp nhau sẽ được phân tách nhau bởi duy nhất 1 dấu cách.
Đầu ra: Đầu ra là giá trị điểm trung bình môn học của 6 môn học, lấy đến 1 chữ số phần thập phân.

```python
9, 8, 7, 9, 10, 7 == map(float, input("9 8 7 9 10 7").split())
tongdiem = (9 + 8 + 7) + (9 + 10) * 2 + 7 * 3
diemtrungbinh = tongdiem / 10
print(f"{diemtrungbinh:.1f}")
```

bai 8: Fahrenheit, hay độ F, là một thang nhiệt độ được đặt theo tên nhà vật lý người Đức Daniel Gabriel Fahrenheit (1686–1736). Dựa vào nhiệt độ Celcius được nhập từ bàn phím, hãy chuyển sang độ F theo công thức: F = 9/5*C + 32 và in ra màn hình kết quả. Kết quả làm tròn đến hai chữ số thập phân đằng sau dấu phẩy.

```python
C = 36
F = (9/5 * C) + 32
print(f"{F:.2f}")
```

bai 9: Sơn Tùng mua một chiếc đồng hồ Movado trên trang mua sắm trực tuyến Amazon. Chiếc đồng hồ có giá x USD (x nhập từ bàn phím). Bởi Amazon không vận chuyển trực tiếp về Việt Nam, Đạt phải trả 10 USD cho phí vận chuyển từ một nước trung gian. Khi vận chuyển về Việt Nam, Đạt được thông báo sẽ phải trả thêm 30% thuế xuất nhập cảnh và 10% thuế VAT cho chiếc đồng hồ nói trên. Viết chương trình tính tổng số tiền mà Đạt phải trả để mua được chiếc đồng hồ trên (USD). Kết quả làm tròn đến hai chữ số thập phân.

```python
x = 100
gia_khi_tinh_phi_van_chuyen = x + 10
gia_khi_tinh_thue_nhap_canh = x * 0.3
gia_khi_tinh_VAT = x * 0.1
tong_so_tien = gia_khi_tinh_phi_van_chuyen + gia_khi_tinh_thue_nhap_canh + gia_khi_tinh_VAT
print(f"{tong_so_tien:.2f}")
```
