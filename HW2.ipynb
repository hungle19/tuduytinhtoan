### : Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(int,input().split())
delta = b**2 - 4* a*c
x1 = (-b + delta** 0.5)/2*a
x2 = (-b - delta** 0.5)/2*a
print(f'x1 ={x1:.2f},x2 ={x2:.2f}')
```

### Giải hệ phương trình tuyến tính:
### ax  by  m
### cx  dy  n

```python
a, b, m, c, d, n = map(int,input().split())
d = a*d - c*b
dx = m*d - n*b
dy = a*n-c*m
x= dx/d
y = dy/d
print(x,y)
```
